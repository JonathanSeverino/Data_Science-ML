```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Alura/Data science/Dados/Aluguel_residencial.csv', sep= ';')
dados.head(10)
```

```python
#Criando váriavel "Valor Bruto"

dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']

dados.head()
```

```python
dados['Valor m²'] = dados['Valor'] / dados['Area']
dados['Valor m²'] = dados['Valor m²'].round(2) #arredondando em 2 casas decimais
dados.head()
```

```python
dados['Valor Bruto m²'] = dados['Valor Bruto'] / dados['Area']
dados['Valor Bruto m²'] = dados['Valor Bruto m²'].round(2)

dados.head(10)
```

```python
#Criando uma váriavel agregando os tipos de imóveis em casa e apartamento

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']

dados['Tipos Agregados'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')

dados.head(10)
```

##Excluindo variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipos Agregados', 'Valor Bruto', 'Valor m²', 'Valor Bruto m²']])

dados_aux.head(10)
```

```python
#Exluindo variável com del

del dados_aux['Valor Bruto']
dados_aux.head()
```

```python
#Exluindo usando o pop

dados_aux.pop('Valor Bruto m²')
```

```python
dados_aux.head()
```

```python
#Exluindo na Base original

dados.drop(["Valor Bruto", "Valor Bruto m²"], axis= 1, inplace= True)
dados.head(10)
```

```python
#Salvando arquivo
dados.to_csv('/content/drive/MyDrive/Alura/Data science/Dados/Aluguel_residencial.csv', sep= ';', index = False)
```
